```python
import pandas as pd
import random

lst = ['robot'] * random.randint(5,15)
lst += ['human'] * random.randint(5,15)
lst += ['sosedka'] * random.randint(5,15)
random.shuffle(lst)
data = pd.DataFrame({'who Am I':lst})
data.head()
```

```python
pd.get_dummies(data['who Am I'])
```

```python
data.loc[data['who Am I'] == 'robot', 'robot'] = 1
data.loc[data['who Am I'] != 'robot', 'robot'] = 0
data.loc[data['who Am I'] == 'human', 'human'] = 1
data.loc[data['who Am I'] != 'human', 'human'] = 0
data.loc[data['who Am I'] == 'sosedka', 'sosedka'] = 1
data.loc[data['who Am I'] != 'sosedka', 'sosedka'] = 0
data[['robot', 'human', 'sosedka']]
```

```python
import pandas as pd 
import random

lst = ['robot'] * random.randint(5,15)
lst += ['human'] * random.randint(5,15)
lst += ['sosedka'] * random.randint(5,15)
random.shuffle(lst)
data = pd.DataFrame({'who Am I': lst})
# print(data)
print()
#------------------------------------------------------#

data['tmp'] = 1
data.set_index([data.index, 'who Am I'], inplace=True)
data = data.unstack(level = 1, fill_value = 0).astype(int)
data.columns = data.columns.droplevel()
data.columns.name = None
print(data)
```
